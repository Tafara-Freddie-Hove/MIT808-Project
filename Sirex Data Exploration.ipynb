```python
#EDA-Exploratory Data Analysis 


### #Load libraries and Dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

#Reading files
sirex = pd.read_csv('C:/Users/tfhove/Desktop/data/sirex_dat.csv')
```

```python
#Checking missing values
sirex.isnull().sum()
```

```python
#Dropping unwanted columns
drop_list = ['Company1','Plantation (Company cad all WGS84.shp)','Plantation (group)','Region1','Species (group)']
sirex = sirex.drop(drop_list, axis = 1)
```

```python
#Drop duplicate rows
sirex = sirex.drop_duplicates()
```

```python
#Check the dimensions
sirex.shape
```

```python
#Handling missing values with fillna method and mode
cols = sirex.columns
sirex[cols]= sirex[cols].fillna(sirex.mode().iloc[0])
sirex.head()
```

```python
#Rename inconsistent column names
sirex = sirex.rename(columns= {'AGE':'Age'})
sirex.head()
```

```python
#Checking missing values
sirex.isnull().sum()
```

```python
#Convert the column names to lower case
sirex['Region']=sirex['Region'].str.lower()
sirex.head()
```

```python
#Cleaning the Region column
sirex['Region'].unique()
```

```python
sirex['Region']=sirex['Region'].replace(['central districts'],'central district')
```

```python
#Cleaning the Species column
sirex['Species'].unique()
```

```python
#Convert the column names to lower case
sirex['Species']=sirex['Species'].str.lower()
```

```python
#remove trailing white space
sirex['Species']=sirex['Species'].str.strip()
sx = sirex['Species'].unique()
sx.sort()
sx
```

```python
#Dealing with non-standard missing values
sirex['Species'] = sirex['Species'].replace(['-','none','na','unknown','n/a',
          '.','unkown','unkown','?', '_','?'],np.nan)

cols = sirex.columns
sirex[cols]=sirex[cols].fillna(sirex.mode().iloc[0])
```

```python
sirex['Species'] = sirex['Species'].replace(['p?','p pat','p pat or p ell?',
 'p pat; p ell','p pat?','p pats','p. pat','ppat','p exc','p ?','ppin'],'pinus patula')

sirex['Species'] = sirex['Species'].replace(['p.rad','p. rad','p rad','prad',],'pinus radiata')

sirex['Species'] = sirex['Species'].replace(['pell','p ell','p. ell','p pell','p. pat/ell','pptl'],'pinus elliottii')

sirex['Species'] = sirex['Species'].replace(['p tae','p. tae','p tea', 'pexc','ptae','p.tae'],'pinus taeda')

sirex['Species'] = sirex['Species'].replace(['p gre','p gres','pgre',
    'p grs ','p. grs','pgrn', 'pgrs','p grs','p. greg s','p gren'], 'pinus greggi')

sirex['Species'] = sirex['Species'].replace(['p mix','p. mix',
                     'pech','pmix'],'pinus maximinoi')
```

```python
# Find common Species
sp = sirex.loc[:,'Species']
sp
```

```python
#Dominant Species
# Plot a bar graph of count vs Species
sns.set_style('whitegrid')
sp.value_counts().plot(kind = 'bar',title='count vs Species type')
```

```python
#Sirex Presence in Species
sns.set_style('whitegrid')
sns.countplot(x = 'Sirex Presence', data = sirex )
plt.title("Sirex Presence in Species")
```

```python
#group data by Sirex Presence to display plot against Region
sirex_reg= sirex.groupby(['Region','Sirex Presence']).size().unstack().fillna(0)
sirex_reg
```

```python
#Plot bar chart showing total complaints Vs City
sirex_reg.sum(axis=1).sort_values(ascending = False).plot(kind='bar',figsize=(6,9))
plt.title("Sirex Presence by Region")
```

```python
total_dead = sirex.pivot_table('Tot dead',index='Year',columns='Sirex Presence',aggfunc=sum)
```

```python
total_dead.head()
```

```python
#The effects of Sirex on Species from 2012 to 2019
total_dead.plot(title="Total Species died of Sirex and Year")
```

```python
#Decomposing the Sirex Presence column into Sirex and No Sirex
sirex['sirex_no.sirex'] = np.where(sirex['Sirex Presence'].str.contains('Sirex'), 'Sirex','No Sirex')
```

```python
sirex['sirex_no.sirex'][:5]
```

```python
#Grouping the data by the Species
by_spicies = sirex.groupby(['Species','sirex_no.sirex'])

agg_counts = by_spicies.size().unstack().fillna(0)
agg_counts[:10]
```

```python
#Select the top Species 
indexer = agg_counts.sum(1).argsort()
indexer[:10]
```

```python
#Select rows with largest values
count_subset = agg_counts.sum(1).nlargest(10)
count_subset
```

```python
#Rearrange the data for plotting
count_subset = count_subset
count_subset.name = 'total'
count_subset = count_subset.reset_index()
count_subset[:10]
```

```python
#Effect of Sirex on Species
sns.barplot(x='total',y='Species', hue = sirex['sirex_no.sirex'], data = count_subset)
plt.title("Total number of Species with Sirex")
```
